```python
#Loading libraries 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (5.0, 4.0)
import seaborn as sns
import xlrd
```

```python
#loading data
train = pd.read_excel("Report1.xlsx")
```

```python
sns.countplot(train.FormNumberMatched)
```

```python
sns.countplot(x="FormNumberMatched",  hue="FormNumberMatched", data=train, )
```

```python
train["FNIsNull"]=train["ActualFormNumber"].isnull()
train["FormNumber"] = train.apply(lambda row: row.ActualFormNumber  if not row.FNIsNull
                                  else row.ScreenFormNumber, axis=1)
#train["ActualFormNumber"] if pd.isnull(train["ActualFormNumber"]) else train["ScreenFormNumber"]
plt.rcParams['figure.figsize'] = (50.0, 8.0)
train.plot(x="FormNumber", y=["FormNumberCountInWord", "NumberOfPages"], kind="bar")
```

```python
#plt.rcParams['figure.figsize'] = (8.0, 8.0)
#train.groupby("FormNumberMatched").count().FormNumber.plot(kind='pie')
explode = (0, 0.1)
train['FormNumberMatched'].value_counts().plot.pie(figsize=(6, 6),explode=explode, autopct='%1.1f%%',
        shadow=True, startangle=90)
```

```python
explode = (0, 0.1)
train['IsFormInPacket'].value_counts().plot.pie(figsize=(6, 6),explode=explode, autopct='%1.1f%%',
        shadow=True, startangle=90)
```
